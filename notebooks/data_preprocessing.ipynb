```python

df = df.dropna(subset=['application', 'documents'])
df = df.reset_index(drop=True)
```

```python
#filling the null values in tags using the other rows of columns as a vocabulary and the other columsn of the empty rows like details , eligibilty etc
existing_tags = df['tags'].dropna().str.lower().str.split(', ')
tag_vocab = sorted(set(tag for sublist in existing_tags for tag in sublist))

kw_model = KeyBERT(model='all-MiniLM-L6-v2')

def get_combined_text(row):
    return f"{row['eligibility']} {row['schemeCategory']}"

def extract_relevant_tags(text, tag_vocab, top_n=5):
    keywords = kw_model.extract_keywords(
        text,
        keyphrase_ngram_range=(1, 2),
        stop_words='english',
        use_maxsum=True,
        nr_candidates=20,
        top_n=top_n
    )

    matched_tags = []
    for word, score in keywords:
        for tag in tag_vocab:
            if tag in word or word in tag:
                matched_tags.append(tag)
                break
    return ', '.join(sorted(set(matched_tags))) if matched_tags else ', '.join(word for word, _ in keywords)

df['tags'] = df.apply(
    lambda row: extract_relevant_tags(get_combined_text(row), tag_vocab)
    if pd.isnull(row['tags']) else row['tags'],
    axis=1
)

```

```python

df.drop_duplicates(inplace=True)
```

```python

df["full_text"] = (
    df["scheme_name"].fillna('') + " | " +
    df["details"].fillna('') + " | Eligibility: " +
    df["eligibility"].fillna('') + " | Benefits: " +
    df["benefits"].fillna('')+
    " | Category: "+ df["schemeCategory"].fillna('')
)
```

```python
def clean_text(text):
    return text.lower().replace('\n', ' ').replace(r"\s+", " ").strip()

df["full_text"] = df["full_text"].apply(clean_text)
```

```python
persist_dir = "chroma_store"
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
df = df.drop_duplicates(subset=["slug", "full_text"])  # avoid redundant docs
docs = [
    Document(
        page_content=row.get("full_text", ""),
        metadata={
            "scheme_name": row.get("scheme_name", ""),
            "slug": row.get("slug", ""),
            "eligibility": row.get("eligibility", ""),
            "benefits": row.get("benefits", ""),
            "documents": row.get("documents", ""),
            "application": row.get("application", ""),
            "level": row.get("level", ""),
            "category": row.get("schemeCategory", ""),
            "tags": row.get("tags", ""),
            "url": f"https://www.myscheme.gov.in/schemes/{row.get('slug', '')}"
        }
    )
    for _, row in df.iterrows()
]
```

```python
from langchain.vectorstores import Chroma

vectorstore = Chroma.from_documents(
    documents=docs,
    embedding=embedding_model,
    persist_directory=persist_dir

)

vectorstore.persist()
```
